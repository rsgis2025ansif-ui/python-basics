## 6. Mini Task â€“ Rainfall classification
### ðŸ”¶ Task 6.1 â€“ Classify rainfall intensity

```python
rain = [0.0, 12.5, 34.0, 2.0, 0.0, 55.1, 8.4]

no_rain = 0
light = 0
moderate = 0
heavy = 0

for r in rain:
 if r == 0:
  no_rain += 1
 elif r > 0 and r <= 10:
  light += 1
 elif r > 10 and r <= 30:
  moderate += 1
else:
  heavy += 1

print("No rain days:", no_rain)
print("Light rain days:", light)
print("Moderate rain days:", moderate)
print("Heavy rain days:", heavy)
```
